```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv("CKD.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
dataset[['classification_yes']].value_counts()
```

```python
indep=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
dep=dataset[['classification_yes']]
```

```python
indep.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(indep,dep,test_size=1/3,random_state=0)
```

```python
from lightgbm import LGBMClassifier
from sklearn.model_selection import GridSearchCV
param_grid={'n_estimators':[100, 200],'learning_rate':[0.01, 0.1],'num_leaves':[31, 50, 100],'max_depth':[-1, 10, 20],'min_child_samples':[20, 50, 100],'subsample':[0.8, 1.0],'colsample_bytree':[0.8, 1.0],'reg_alpha':[0, 0.1, 1],'reg_lambda':[0, 0.1, 1]}
grid=GridSearchCV(LGBMClassifier(),param_grid,refit=True,verbose=3,n_jobs=-1,scoring='f1_weighted')
grid.fit(x_train,y_train)
```

```python
re=grid.cv_results_
y_pred=grid.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,f1_score,roc_auc_score
cm=confusion_matrix(y_test,y_pred)
clf_report=classification_report(y_test,y_pred)
f1_macro=f1_score(y_test,y_pred,average='weighted')
roc_auc_score(y_test,grid.predict_proba(x_test)[:,1])
```

```python
print("The Confusion Matrix:\n",cm)
```

```python
print("The Report:\n",clf_report)
```

```python
print("The f1_macro value for the best parameter{}:".format(grid.best_params_),f1_macro)
```
